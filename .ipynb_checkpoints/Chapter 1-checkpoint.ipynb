## Code for Sketches

```python
import numpy as np
import matplotlib.pyplot as plt

def vf_plot(url, section, coords="r", fx=lambda x,y,z: 0, fy=lambda x,y,z: 0, fz=lambda x,y,z: 0, title="", length=0.05):
    if coords == "r":
        r = np.linspace(0, 12, 4)
        t = np.linspace(0, 2 * np.pi, 16, endpoint=False)
        z = np.linspace(-12, 12, 3)
    
        R, T, Z = np.meshgrid(r, t, z, indexing="ij")
        X = R * np.cos(T)
        Y = R * np.sin(T)
        U = fx(X, Y, Z)
        V = fy(X, Y, Z)
        W = fz(X, Y, Z)

        fig = plt.figure(figsize=(6, 6))
        ax = fig.add_subplot(111, projection="3d")
        ax.quiver(X, Y, Z, U, V, W, length=length, normalize=False, color="blue", arrow_length_ratio=0.3)
    
        ax.set_title(title, fontsize=16)
        ax.set_xlabel("x", fontsize=14)
        ax.set_ylabel("y", fontsize=14)
        ax.set_zlabel("z", fontsize=14)
        ax.set_box_aspect([1, 1, 1])
        ax.tick_params(labelsize=12)

        plt.tight_layout()
        plt.savefig(f"sketches/{section}/vector_field_{url}.png", dpi=600, bbox_inches="tight")
        plt.show()
    else:
        x = np.linspace(-8, 8, 4)
        y = np.linspace(-8, 8, 4)
        z = np.linspace(-8, 8, 4)

        X, Y, Z = np.meshgrid(x, y, z)
        
        U = fx(X, Y, Z)
        V = fy(X, Y, Z)
        W = fz(X, Y, Z)

        fig = plt.figure(figsize=(6, 6))
        ax = fig.add_subplot(111, projection="3d")
        ax.quiver(X, Y, Z, U, V, W, length=length, normalize=False, color="blue", arrow_length_ratio=0.3)
    
        ax.set_title(title, fontsize=16)
        ax.set_xlabel("x", fontsize=14)
        ax.set_ylabel("y", fontsize=14)
        ax.set_zlabel("z", fontsize=14)
        ax.set_box_aspect([1, 1, 1])
        ax.tick_params(labelsize=12)
    
        plt.tight_layout()
        plt.savefig(f"sketches/vector_field_{url}.png", dpi=600, bbox_inches="tight")
        plt.show()

    
```

#### 1.1.5 Using sum notation (discussed in Section 0.1), write the vectors in the margin at left as a sum of multiples of standard basis vectors.

$\begin{bmatrix}1\\.\\.\\.\\1\\1\end{bmatrix}$,    $\begin{bmatrix}1\\2\\.\\.\\.\\n - 1\\n\end{bmatrix}$,    $\begin{bmatrix}0\\0\\3\\4\\.\\.\\.\\n - 1\\n\end{bmatrix}$

#### 1.1.7 Sketch the following vector fields:

a. $\vec{F}\begin{pmatrix}x\\y\\z\end{pmatrix} = \begin{bmatrix}0\\0\\x^2 + y^2\end{bmatrix}$ &emsp; b. $\vec{F}\begin{pmatrix}x\\y\\z\end{pmatrix} = \begin{bmatrix}y\\-x\\-z\end{bmatrix}$ &emsp; c. $\vec{F}\begin{pmatrix}x\\y\\z\end{pmatrix} = \begin{bmatrix}x + y\\x - y\\-z\end{bmatrix}$

#### Solution
a.

```python
vf_plot("a", "1.1", fz=lambda x, y, z: x**2 + y**2, title=r"$\vec{F}(x, y, z) = \langle 0, 0, x^2 + y^2 \rangle$", length=0.025)
```

b.

```python
vf_plot("b", "1.1", fx=lambda x,y,z: y, fy=lambda x,y,z: -x, fz=lambda x,y,z: -z, length=0.1, title=r"$\vec{F}(x, y, z) = \langle y, -x, -z \rangle$")
```

c.

```python
vf_plot("c", "1.1", fx=lambda x,y,z: x + y, fy=lambda x,y,z: x - y, fz=lambda x,y,z: -z, length=0.1, title=r"$\vec{F}(x, y, z) = \langle x + y, x - y, -z \rangle$")
```

#### 1.1.9 Show that the set of complex numbers $\{z|Re(wz) = 0\}$ with fixed $w \in \mathbb{C}$ is a subspace of $\mathbb{R^2} = \mathbb{C}$. Describe this subspace.
<br><br>
#### Solution

1. contains 0

2. closed under addition:
<br>&emsp; if $a,b \in z$
<br>&emsp; $Re(w(a + b)) = Re(wa) + Re(wb) = 0$

4. Closed under multiplication:
<br>&emsp; if $a \in z$ and $c \in \mathbb{C}$
<br>&emsp; $Re(wca) = cRe(wa) = 0$

#### 1.2.3 Given the matrices A and B,

$A = \begin{bmatrix}1 & 2 & 0\\3 & 1 & -1\end{bmatrix}$ &emsp; $B = \begin{bmatrix}2 & 5 & 1\\1 & 4 & 2\\1 & 3 & 3\end{bmatrix}$

a. compute the third column of AB without computing the entire matrix AB.<br>
b. compute the second row of AB, again without computing the entire matrix

<br><br>
#### Solution

a. $AB_:,3 = \begin{bmatrix}A_1 \cdot B_3\\A_2 \cdot B_3\end{bmatrix}$<br><br>
&emsp; $= \begin{bmatrix}\begin{bmatrix}1\\2\\0\end{bmatrix} \cdot \begin{bmatrix}1\\2\\3\end{bmatrix}\\\begin{bmatrix}3\\1\\-1\end{bmatrix} \cdot \begin{bmatrix}1\\2\\3\end{bmatrix}\end{bmatrix}$

&emsp; $= \begin{bmatrix}1 \cdot 1 + 2 \cdot 2 + 0 \cdot 3\\3 \cdot 1 + 1 \cdot 2 -1 \cdot 3\end{bmatrix}$ 

<br><br>
$= \begin{bmatrix}5\\2\end{bmatrix}$

<br>

---

b. $AB_2 = \begin{bmatrix}A_2 \cdot B_1 & A_2 \cdot B_2 & A_2 & B_3\end{bmatrix}$
$= \begin{bmatrix}\begin{bmatrix}3\\1\\-1\end{bmatrix} \cdot \begin{bmatrix}2\\1\\1\end{bmatrix} & \begin{bmatrix}3\\1\\-1\end{bmatrix} \cdot \begin{bmatrix}5\\4\\3\end{bmatrix} & \begin{bmatrix}3\\1\\-1\end{bmatrix} \cdot \begin{bmatrix}1\\2\\3\end{bmatrix}\end{bmatrix}$

$= \begin{bmatrix}6 + 1 - 1 & 15 + 4 - 3 & 3 + 2 - 3\end{bmatrix}$

$= \begin{bmatrix}6 & 16 & 2\end{bmatrix}$

#### 1.2.9 Given the two matrices

$A = \begin{bmatrix}1 & 0\\1 & 0\end{bmatrix}$ &emsp; and &emsp; $B = \begin{bmatrix}1 & 0 & 1\\2 & 1 & 0\end{bmatrix}$,

a. What are their transposes?<br>
b. Without computing $AB$, what is $(AB)^\top$?<br>
c. Confirm your result by computing $AB$.<br>
d. What happens if you do part b using the _incorrect_ formula $(AB)^\top = A^\top B^\top$?

#### Solution

a. $A^\top = \begin{bmatrix}1 & 1\\0 & 0\end{bmatrix} $ &emsp; and &emsp; $B^\top = \begin{bmatrix}1 & 2\\0 & 1\\1 & 0\end{bmatrix}$

b. $(AB)^\top = B^\top A^\top$<br><br>
$= \begin{bmatrix}1 & 2\\0 & 1\\1 & 0\end{bmatrix}\begin{bmatrix}1 & 1\\0 & 0\end{bmatrix}$

$= \begin{bmatrix}\begin{bmatrix}1\\2\end{bmatrix}\begin{bmatrix}1\\0\end{bmatrix} & \begin{bmatrix}1\\2\end{bmatrix} \begin{bmatrix}1\\0\end{bmatrix}\\\begin{bmatrix}0\\1\end{bmatrix}\begin{bmatrix}1\\0\end{bmatrix} & \begin{bmatrix}0\\1\end{bmatrix}\begin{bmatrix}1\\0\end{bmatrix}\\\begin{bmatrix}1\\0\end{bmatrix}\begin{bmatrix}1\\0\end{bmatrix} & \begin{bmatrix}1\\0\end{bmatrix}\begin{bmatrix}1\\0\end{bmatrix}\end{bmatrix}$

$= \begin{bmatrix}1 \cdot 1 + 2 \cdot 0 & 1 \cdot 1 + 2 \cdot 0\\0 \cdot 1 + 1 \cdot 0 & 0 \cdot 1 + 1 \cdot 0\\1 \cdot 1 + 0 \cdot 0 & 1 \cdot 1 + 0 \cdot 0\end{bmatrix}$

$= \begin{bmatrix}1 & 1\\0 & 0\\1 & 1\end{bmatrix}$

c. $(AB)^\top = \begin{pmatrix}\begin{bmatrix}1 & 0\\1 & 0\end{bmatrix}\begin{bmatrix}1 & 0 & 1\\2 & 1 & 0\end{bmatrix}\end{pmatrix}^\top$

$= \begin{pmatrix}\begin{bmatrix}1 \cdot 1 + 0 \cdot 2 & 1 \cdot 0 + 0 \cdot 1 & 1 \cdot 1 + 0 \cdot 0\\1 \cdot 1 + 0 \cdot 2 & 1 \cdot 0 + 0 \cdot 1 & 1 \cdot 1 + 0 \cdot 0\end{bmatrix}\end{pmatrix}^\top$

$= \begin{pmatrix}\begin{bmatrix}1 & 0 & 1\\1 & 0 & 1\end{bmatrix}\end{pmatrix}^\top$

$= \begin{bmatrix}1 & 1\\0 & 0\\1 & 1\end{bmatrix}$

d. $(AB)^\top = \begin{bmatrix}1 & 1\\0 & 0\end{bmatrix}\begin{bmatrix}1 & 2\\0 & 1\\1 & 0\end{bmatrix}$ &emsp; (incorrect formula applied)

Impossible, since A's dimensions are 2 x 2 and B's dimensions are 3 x 2.
